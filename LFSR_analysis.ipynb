```python
from numpy import mat,transpose
from numpy.linalg import solve,inv
```

```python
def str_xor(my_str1,my_str2):
    '''字符串异或操作'''
    res = ""
    for i in range(0,len(my_str1)):
        #变成10进制是转化成字符串 2进制与10进制异或结果一样，都是1,0
        xor_res = int(my_str1[i],10)^int(my_str2[i],10) 
        if xor_res == 1:
            res += '1'
        if xor_res == 0:
            res += '0'
    return res
```

```python
def LFSR_de():
    num_of_reg = 5
    M = '011001111111001'
    C = '101101011110011'
    if len(M)!=len(C) or len(M)<2*num_of_reg:
        str = 'input error'
        print(str)
        return
    M_test = M[:2*num_of_reg]
    C_test = C[:2*num_of_reg]
    K = str_xor(M_test,C_test)
    key_list = []
    for i in range(num_of_reg):
        key_list.append([int(K[j]) for j in range(i,i+num_of_reg)])
#     用numpy库构造两个矩阵
    mat_key = mat(key_list)
    mat_key2 = transpose(mat([int(k) for k in K[num_of_reg:]]))
#     用numpy库解一元线性方程
    feedback_key = solve(mat_key,mat_key2).astype(int)
    for i in feedback_key:
        k =str(abs(i))
        print(k[2],end = '')

```

```python
LFSR_de()
```
