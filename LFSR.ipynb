```python
def str2bin(message):
    '''字符串转二进制'''
    res = ""
    for i in message:
        tmp = bin(ord(i))[2:]
        for j in range(0,8-len(tmp)):
            tmp = '0'+ tmp   #把输出的b给去掉
        res += tmp
    return res


def str_xor(my_str1,my_str2):
    '''字符串异或操作'''
    res = ""
    for i in range(0,len(my_str1)):
        #变成10进制是转化成字符串 2进制与10进制异或结果一样，都是1,0
        xor_res = int(my_str1[i],10)^int(my_str2[i],10) 
        if xor_res == 1:
            res += '1'
        if xor_res == 0:
            res += '0'
    return res
```

```python
def getindex(fb_key):
    '''获取抽头位'''
    fb_list = []
    for i in range(len(fb_key)):
        if fb_key[i] == '1':
            fb_list.append(i)
    return fb_list
def fb_fun(fb_list,cr_list):
    '''反馈函数'''
    res = cr_list[fb_list[0]]
    for i in range(len(fb_list)-1):
        res = str_xor(res,cr_list[fb_list[i+1]])
    return res
```

```python
def LFSR(num_regs,fb_key,init_key):
    '''LFSR过程'''
#     参数初始化
    count = 0 # 计数器
    cycle = 0 # 周期
    seq = '' # 输出序列
    old_seq=[] # 旧状态列表
    cr_list = init_key
    fb_list = getindex(fb_key) # 用列表保存抽头位
    print('初态:',init_key)
    while True:
#     取出抽头位
        count+=1
        seq+=cr_list[-1]
        print('---------第%d轮---------'%count)
        print('本轮输出序列:',seq)
#     反馈函数fb_fun计算,更新状态,同时保存当前序列以判断周期
        old_seq.append(cr_list)
        print(old_seq)
        cr_list = fb_fun(fb_list,cr_list)+ cr_list[0:-1]
        print('更新后的状态:',cr_list)
#     判断并计算周期
        for i in range(len(old_seq)):
#         如果出现相同状态，则完成一轮循环
            if cr_list == old_seq[i]:
                cycle = len(old_seq)-i
                return seq,cycle
                
```

```python
def menu():
    '''输入和初始化'''
    while True:
        print('please input the num of regs:')
        num_regs = int(input())
        print('please input the feedback key(bin):')
        fb_key = input()
        print('please input the init key(bin):')
        init_key = input()
        if num_regs == len(fb_key) == len(init_key):
            break
        else:
            print("lenth dosen't match!")
    return num_regs,fb_key,init_key
```

```python
def main():
    num_regs,fb_key,init_key = menu()
    seq,cycle = LFSR(num_regs,fb_key,init_key)
    print("ouput seq: ",seq)
    print("cycle: ",cycle)
main()
```
